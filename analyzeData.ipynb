```python
score_labels = []
with open(score_file) as sf:
    for l in sf:
        if 'score' in l:
            for s in l.split():
                score_labels.append(s)
```

```python
scores = {}
with open(score_file) as sf:
    for l in sf:
        if 'score' not in l:
            for i,s in enumerate(l.split()):
                if score_labels[i] != 'SCORE:':
                    if score_labels[i] not in scores:
                        scores[score_labels[i]] = []
                    if score_labels[i] == 'description':
                        scores[score_labels[i]].append(s)
                    else:
                        scores[score_labels[i]].append(float(s))
```

```python
for s in scores:
    scores[s] = np.array(scores[s])
```

```python
def readRosettaScoreFile(score_file):
    
    score_labels = []
    with open(score_file) as sf:
        for l in sf:
            if 'score' in l:
                for s in l.split():
                    score_labels.append(s)
    
    scores = {}
    with open(score_file) as sf:
        for l in sf:
            if 'score' not in l:
                for i,s in enumerate(l.split()):
                    if score_labels[i] != 'SCORE:':
                        if score_labels[i] not in scores:
                            scores[score_labels[i]] = []
                        if score_labels[i] == 'description':
                            scores[score_labels[i]].append(s)
                        else:
                            scores[score_labels[i]].append(float(s))
    return scores
```

```python
def sigmoid(x, x0, m, w=1.0):
    return w*(1/(1+np.exp(-m*( x-x0 ))) - 0.5 - 0.5)
```

```python
def bounded(x, lb, ub, sd, rswitch, w=1.0):
    B = []
    for xi in x:
        if xi < lb:
            y = ((xi-lb)/sd)**2
        elif xi >= lb and xi <= ub: 
            y = 0
        elif xi > ub and xi <= ub +rswitch*sd:
            y = ((xi-ub)/sd)**2
        elif xi > ub +rswitch*sd:
            y = (1/sd)*(xi-(ub+rswitch*sd))+(rswitch*sd/sd)**2
        B.append(y)
    B = w*np.array(B)
    return B
```

```python
# Create ploting axes
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212, sharex=ax1)

# Plot sigma function
d0 = 6.135
delta = 10
x = np.arange(d0-delta,d0+delta,0.01)
ax1.plot(x, sigmoid(x, d0, 5.181, w=3.140))
ax1.set_xlabel('x')
ax1.set_ylabel('Sigmoid(x)')


# Plot Bounded function
ub = 6.328
delta = 10
x = np.arange(ub-delta,ub+delta,0.01)
ax2.plot(x, bounded(x, 0, ub, 1, 0.5, w=3.140))
ax2.set_xlabel('x')
ax2.set_ylabel('Bounded(x)')
```

```python
nocst_score = readRosettaScoreFile('02.runAbinitioPrediction/output/153L_A_abinitio.sc')
bounded_score = readRosettaScoreFile('05.runCstAbinitioPrediction/output/153L_abinitio_bounded.sc')
sigmoid_score = readRosettaScoreFile('05.runCstAbinitioPrediction/output/153L_abinitio_sigmoid.sc')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132, sharex=ax1, sharey=ax1)
ax3 = fig.add_subplot(133, sharex=ax1, sharey=ax1)

ax1.scatter(nocst_score['rms'], nocst_score['score'], label='nocst', alpha=0.5)
ax2.scatter(sigmoid_score['rms'], sigmoid_score['score'], label='sigmoid', alpha=0.5)
ax3.scatter(bounded_score['rms'], bounded_score['score'], label='bounded', alpha=0.5)

ax1.errorbar(np.average(nocst_score['rms']), 
             np.average(nocst_score['score']), 
             xerr=np.std(nocst_score['rms']),
             yerr=np.std(nocst_score['score']),
             c='k', capsize=3)

ax2.errorbar(np.average(sigmoid_score['rms']), 
             np.average(sigmoid_score['score']), 
             xerr=np.std(sigmoid_score['rms']),
             yerr=np.std(sigmoid_score['score']),
             c='k', capsize=3)

ax3.errorbar(np.average(bounded_score['rms']), 
             np.average(bounded_score['score']), 
             xerr=np.std(bounded_score['rms']),
             yerr=np.std(bounded_score['score']),
             c='k', capsize=3)

print('Nocst average RMSD: %.2f' % np.average(nocst_score['rms']))
print('Sigma average RMSD: %.2f' % np.average(sigmoid_score['rms']))
print('Bounded average RMSD: %.2f' % np.average(bounded_score['rms']))


ax1.set_title('nocst')
ax2.set_title('sigmoid')
ax3.set_title('bounded')

ax2.get_yaxis().set_visible(False)
ax3.get_yaxis().set_visible(False)

ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['left'].set_visible(False)

ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax3.spines['left'].set_visible(False)

ax2.set_xlabel('RMSD to native structure [Angstroms]')
ax1.set_ylabel('Potential energy [kcal/mol]')


plt.ylim(-500, 0)
```
